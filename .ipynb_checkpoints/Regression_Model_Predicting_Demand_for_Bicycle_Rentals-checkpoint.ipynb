### Define X and y

```python
##log transform the target variable
train["count_log"] = np.log1p(train["count"])
```

```python
y = train["count_log"]
```

```python
X = train[['season', 'workingday', 'weather', 'temp', 'year', 'hour', 'humidity']]
```

### Split data to train-test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X,y, test_size = 0.2, random_state = 40) 
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

### Post-split Feature Engineering

```python
# instantiate our KBinsDiscretizer object
binner = KBinsDiscretizer(n_bins = 8, strategy = 'uniform')
```

```python
# fit the binner to our data and perform the transformation
# then convert the sparse matrix to an array
enc_bins = binner.fit_transform(train[['hour']]).toarray()
```

```python
enc_bins
```

```python
# define transformers - name, method, target
transformers= [('ohe', OneHotEncoder(drop = 'first', 
                        handle_unknown = 'ignore'), ['year', 'season', 'weather', 'workingday', 'hour']),
                ('scaler', RobustScaler(), ['temp', 'humidity'])]
```

```python
# now we instantiate our ColumnTransformer() object
column_transformer = ColumnTransformer(transformers, 
                                       remainder = 'drop')
```

```python
# fit and transform the ColumnTransformer() (ONLY using train data)
X_train_fe = column_transformer.fit_transform(X_train, y_train).toarray()
```

```python
X_train_fe = pd.DataFrame(X_train_fe)
```

```python
X_train_fe.head(3)
```

```python
# use the fitted transformer on X_test
X_test_fe = column_transformer.transform(X_test)
```

##  the LR model

```python
m_lr = LinearRegression()
```

```python
m_lr.fit(X_train_fe, y_train)
```

```python
y_pred = m_lr.predict(X_test_fe)
```

```python
# Test score
test_score = round(m_lr.score(X_test_fe, y_test),3)
test_score
```

### model Regularization

```python
# import Ridge and Lasso
import time
from sklearn.linear_model import Ridge, Lasso
from IPython import display
```

```python
train_errors = []
test_errors = []
```

```python
alphas = range(-10, 10, 1)
for alpha in alphas:
    # Create ridge model
    model = Ridge(alpha=alpha)

    # Fit the model
    model.fit(X_train_fe, y_train)
    
    # Calculate the scores
    train_score = model.score(X_train_fe, y_train).round(2)
    test_score = model.score(X_test_fe, y_test).round(2)

    # Calculate test and train predictions
    y_pred_train = model.predict(X_train_fe)
    y_pred_test = model.predict(X_test_fe)

    cv = cross_validate(
    estimator=model, # model
    X=X_train_fe,        # Input features
    y=y_train,        # target data
    cv=5,            # no. of datasets in cross-validation
    scoring="neg_mean_squared_error", # Evaluation metric (MSE)
    return_train_score=True # return both cv-training and cv-test (validation) scores
)
# Create a dataframe
    cv = pd.DataFrame(cv)
    
    # Calculate average mean and test score
    train_error = abs(cv["train_score"].mean())
    test_error = abs(cv["test_score"].mean())
    
    # Append the scores to the list
    train_errors.append(train_error)
    test_errors.append(test_error)
    
    # Make scatter
    ax=sns.scatterplot(x=X_train["temp"], y=y_train)
    ax.set(title=f"alpha={alpha}")
    sns.scatterplot(x=X_test["temp"], y=y_pred_test)
    plt.text(-2.5, 6, f"train score={train_score}")
    plt.text(-2.5, 5, f"test score={test_score}")
    
    # Display properties
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(0.3)
    plt.close()
```

```python
plt.figure(figsize=(15, 8))
ax = sns.lineplot(x=alphas, y=train_errors, color="red", label="train")
sns.lineplot(x=alphas, y=test_errors, color="blue", label="test")
ax.set(xlabel='alphas', ylabel='Mean score')
```
