### Resampling the data to daily and hourly means

```python
data.head(2)
```

```python
# resampling using resample() function 
daily_averages = data.resample('D').mean()
hourly_averages = data.resample('H').mean()
daily_averages['daily_av'] = daily_averages['power']
daily_averages.drop('power', axis=1, inplace=True)
hourly_averages['hourly_av'] = hourly_averages['power']
hourly_averages.drop('power', axis=1, inplace=True)
```

```python
# a 2nd resampling option could be- adding time features and resample on new features
data['year'] = data.index.year
data['month'] = data.index.month
data['hour'] = data.index.hour
data['day'] = data.index.day
data_hourly_grouped = data.groupby(['year', 'month', 'day', 'hour']).mean().reset_index()
data_daily_grouped = data.groupby(['year', 'month', 'day']).mean().reset_index()
data_hourly_grouped['hourly_av'] = data_hourly_grouped['power']
data_hourly_grouped.drop('power', axis=1, inplace=True)
data_daily_grouped['daily_av'] = data_daily_grouped['power']
data_daily_grouped.drop(['power', 'hour'], axis=1, inplace=True)
```

```python
daily_averages['daily_av'].plot(figsize = (20, 6)).set_title("power generated over time");
```

```python
hourly_averages.hourly_av.plot(figsize = (20, 6)).set_title("power generated over time");
```

## Final checks on the data

### Rolling statistics
- a great way to visualize how the dataset is trending. 
- rolling std shows how much the data varies from the rolling average

```python
#Determine rolling statistics
hourly_averages["rolling_avg"] = hourly_averages["hourly_av"].rolling(window=200).mean() #window size 12 denotes 12 months, giving rolling mean at yearly level
hourly_averages["rolling_std"] = hourly_averages["hourly_av"].rolling(window=200).std()

#Plot rolling statistics
plt.figure(figsize=(15,7))
#plt.plot(hourly_averages["hourly_av"], color='#379BDB', label='Original')
plt.plot(hourly_averages["rolling_avg"], color='#D22A0D', label='Rolling Mean')
plt.plot(hourly_averages["rolling_std"], color='#142039', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

```python
#Determine rolling statistics
daily_averages["rolling_avg"] = daily_averages["daily_av"].rolling(window=31).mean() #window size 12 denotes 12 months, giving rolling mean at yearly level
daily_averages["rolling_std"] = daily_averages["daily_av"].rolling(window=31).std()

#Plot rolling statistics
plt.figure(figsize=(15,7))
plt.plot(daily_averages["daily_av"], color='#379BDB', label='Original')
plt.plot(daily_averages["rolling_avg"], color='#D22A0D', label='Rolling Mean')
plt.plot(daily_averages["rolling_std"], color='#142039', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

### **Stationarity test**
Augmented Dickey-Fuller Test -a stationarity test for time series
- Ho: It is non stationary
- H1: It is stationary

```python
import statsmodels
from statsmodels.tsa import stattools
```

```python
def print_adf(data):
    
    """ Prints the results of the augmented Dickey Fuller Test"""
    
    adf_stats, p, used_lag, n_obs, levels, information_criterion = \
            statsmodels.tsa.stattools.adfuller(data)
    
    print(f"""   
            adf: {round(adf_stats,2)}
            p: {p}  
            used lag: {used_lag}
            number of observations: {n_obs}
            {round(levels['1%'],2)} (p=0.01; CI=99%) 
            {round(levels['5%'],2)} (p=0.05; CI=95%)
            {round(levels['10%'],2)} (p=0.1 CI=90%) 
            information criterion (AIC): {information_criterion}
            """)
```

```python
print_adf(daily_averages["daily_av"])
```

```python
print_adf(hourly_averages['hourly_av'])
```

- The ADF test above shows a significant p value, which means we could reject the null hypothesis that the data is not stationary. However we know from the graphs before that there are levels of seasonality in the data.
